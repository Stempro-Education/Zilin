```python
def getContent(url):
    words=[]
    r=requests.get(url, headers={"User-Agent":"Mozilla/5.0"})
    return r.text
```

```python
print(getContent("https://www.nme.com/features/greta-thunberg-full-speech-to-the-un-2019-climate-change-summit-2550824"))
```

```python
import string
def getAllWordsWithOutDup (url):
    soup = BeautifulSoup(getContent(url), 'html.parser')
    words=[]

    for script in soup(["script", "style"]):
        script.extract()
        
    for p in soup.find_all("p"):
        text=p.text.strip()
        for word in text.split():
            temp=word.translate(str.maketrans('', '', string.punctuation)).lower()
            if temp not in words:
                words.append(temp)
    return words

def getAllWordsWithDup (url):
    soup = BeautifulSoup(getContent(url), 'html.parser')
    words=[]

    for script in soup(["script", "style"]):
        script.extract()
        
    for p in soup.find_all("p"):
        text=p.text.strip()
        for word in text.split():
            temp=word.translate(str.maketrans('', '', string.punctuation)).lower()
            words.append(temp)
    return words
```

```python
url="https://www.nme.com/features/greta-thunberg-full-speech-to-the-un-2019-climate-change-summit-2550824"
print(getAllWordsWithOutDup(url))
```

```python
def getWordFreq(word, url):
    return getAllWordsWithDup(url).count(word)
```

```python
url="https://www.nme.com/features/greta-thunberg-full-speech-to-the-un-2019-climate-change-summit-2550824"
print(getWordFreq("i", url))
```

```python
def freqSort(words):
    result = sorted(words, key = words.count, reverse = True)
    return result
```

```python
from nltk.stem import PorterStemmer
from nltk.stem import LancasterStemmer
def Lemma(words):
    porter = PorterStemmer()
    lancaster=LancasterStemmer()
    lemmatize=[]
    for word in words:
        lemmatize.append(porter.stem(word))
    return lemmatize       
    
```

```python
print(Lemma(getAllWordsWithOutDup(url)))
```

```python
def Difficult(words):
    synonym=[]
    antonym=[]
    for word in words:
         for synset in wordnet.synsets(word):
            for lemma in synset.lemmas():
                synonym.append(lemma.name())
            if lemma.antonyms():    
                antonym.append(lemma.antonyms()[0].name())
    return synonym, antonym
```

```python
print(Difficult(getAllWordsWithOutDup(url)))
```
