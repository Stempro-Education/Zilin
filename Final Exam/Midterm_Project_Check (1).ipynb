```python
def extractHeader(filename):
    info=open(filename)
    title=[]
    text=BeautifulSoup(info, 'html.parser')
    for head in text.find_all("thead"):  
        for i in head.find_all("tr"):
             for j in i.find_all("th"):
                title.append(j.text)        
    return title
```

```python
extractHeader("/Users/panyueli/HW/Zilin/Final Exam/2.html")
```

```python
def extractBody(filename):
    info=open(filename)
    bodies=[]
    data=[]
    text=BeautifulSoup(info, 'html.parser')
    for body in text.find_all("tbody"):  
        for i in body.find_all("tr"):
            for j in i.find_all("td"):
                for k in j.find_all("div", class_="show-for-medium-up"):
                    data.append(k.text.replace("\n", "").replace("Save to My Schools","").replace("                    ", ""))
            bodies.append(data)
            data=[]
    return bodies     

```

```python
extractBody("/Users/panyueli/HW/Zilin/Final Exam/2.html")
```

```python
def run(filename):
    result = pd.DataFrame(extractBody(filename), columns=extractHeader(filename))
    return result

    
```

```python
run("/Users/panyueli/HW/Zilin/Final Exam/2.html")
```

```python
def save(htmlfilename, destination):
    run(htmlfilename).to_csv(destination, index=False)
```
